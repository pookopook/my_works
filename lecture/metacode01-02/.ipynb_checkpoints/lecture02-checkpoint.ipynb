## ANOVA 함수

```python

import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
# x : 비교대상 그룹(매장, 온라인, 아울렛) -> sales Method
# y : 비교할 변수 -> 영업이익, 단가
x = 'Sales Method'  #-> 띄여쓰기 컬럼은 ols가 인식못함
y = 'Operating_Margin_rate'

# 띄어쓰기 바꾸기
df_anova = df[[x,y]]
x = x.replace(' ', '_')
y = y.replace(' ', '_')
df_anova.columns = [x,y]
df_anova
```

```python

# C에는 목록이 들어가야 함. cat cols
model = ols(f'{y} ~ C({x})', df_anova).fit

anova_lm(model)

```
